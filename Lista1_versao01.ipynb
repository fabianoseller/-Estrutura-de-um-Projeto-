#### **Exercício 1**

Se uma amostra aleatória n=25, tem uma média amostral de 51,3 e uma desvio padrão populacional de σ=2.

Construa o intervalo com 95% de confiança para a média populacional µ.

**Como o desvio padrão populacional é conhecido e a amostra é aleatória, podemos usar a distribuição normal padrão (z) para calcular o intervalo de confiança da média populacional.**

```r
cat("--- Exercício 1 ---\n")
x_bar_1 <- 51.3
sigma_1 <- 2
n_1 <- 25
conf_level_1 <- 0.95
alpha_1 <- 1 - conf_level_1
z_alpha_2_1 <- qnorm(1 - alpha_1 / 2)
margin_error_1 <- z_alpha_2_1 * (sigma_1 / sqrt(n_1))
ci_lower_1 <- x_bar_1 - margin_error_1
ci_upper_1 <- x_bar_1 + margin_error_1

cat(sprintf("Intervalo de Confiança (95%%): (%.3f, %.3f)\n\n", ci_lower_1, ci_upper_1))

# Gráfico
# Como não temos os dados brutos, representamos a média e o IC
df_ex1 <- data.frame(
  Medida = "Média Populacional",
  Estimativa = x_bar_1,
  Lower = ci_lower_1,
  Upper = ci_upper_1
)

ggplot(df_ex1, aes(x = Medida, y = Estimativa)) +
  geom_point(size = 4, color = "darkblue") +
  geom_errorbar(aes(ymin = Lower, ymax = Upper), width = 0.2, color = "darkblue", linewidth = 1.2) +
  geom_hline(yintercept = x_bar_1, linetype = "dashed", color = "red") +
  labs(title = "Exercício 1: Intervalo de Confiança para a Média (95%)",
       y = "Valor", x = "") +
  theme_minimal() +
  coord_cartesian(ylim = c(ci_lower_1 - 0.5, ci_upper_1 + 0.5))
```

#### **Exercício 3**

Qual deve ser o tamanho da amostra para que o intervalo com 99,5% de confiança para a média populacional tenha uma semi amplitude não superior a 1,5? Sabe-se que a variância populacional é de 23.

**Como a variância populacional é conhecida podemos usar a distribuição normal padrão (z) para determinar o tamanho mínimo necessário da amostra**

```r
cat("--- Exercício 3 ---\n")
E_3 <- 1.5
sigma_sq_3 <- 23
sigma_3 <- sqrt(sigma_sq_3)
conf_level_3 <- 0.995
alpha_3 <- 1 - conf_level_3
z_alpha_2_3 <- qnorm(1 - alpha_3 / 2)
n_3 <- ((z_alpha_2_3 * sigma_3) / E_3)^2
n_3_rounded <- ceiling(n_3)
cat(sprintf("Tamanho da amostra necessário: %.0f\n\n", n_3_rounded))
```

#### **Exercício 4**

Calcular o intervalo de confiança de 95% para a seguinte amostra, com variância populacional desconhecida:   19,8  18,5 17,6 16,7 15,8 15,4 14,1 13,6 11,9 11,4 11,4 8,8 7,5 15,4 15,4 19,5 14,9 12,7 11,9 11,4 10,1 7,9




**Como a variância populacional é desconhecida e a amostra é pequena, utilizamos a distribuição t de Student para calcular o intervalo de confiança de 95% para a média populacional**

```r
cat("--- Exercício 4 ---\n")
# Definir os dados da amostra
sample_data_4 <- c(19.8, 18.5, 17.6, 16.7, 15.8, 15.4, 14.1, 13.6, 11.9, 11.4,
                   11.4, 8.8, 7.5, 15.4, 15.4, 19.5, 14.9, 12.7, 11.9, 11.4,
                   10.1, 7.9)

# Tamanho da amostra
n_4 <- length(sample_data_4)

# Média da amostra
x_bar_4 <- mean(sample_data_4)

# Desvio padrão da amostra
s_4 <- sd(sample_data_4)

# Nível de confiança e erro padrão
conf_level <- 0.95
alpha <- 1 - conf_level
t_critical <- qt(1 - alpha/2, df = n_4 - 1)
se_4 <- s_4 / sqrt(n_4)

# Intervalo de confiança (IC)
ci_lower_4 <- x_bar_4 - t_critical * se_4
ci_upper_4 <- x_bar_4 + t_critical * se_4

cat(sprintf("Intervalo de Confiança (95%%): (%.3f, %.3f)\n\n", ci_lower_4, ci_upper_4))

# Dataframes para plotagem
df_ex4_data <- data.frame(Valor = sample_data_4)
df_ex4_ci <- data.frame(
  Estimativa = x_bar_4,
  Lower = ci_lower_4,
  Upper = ci_upper_4
)

# Gráfico
ggplot() +
  geom_histogram(data = df_ex4_data, aes(x = Valor, y = after_stat(density)),
                 binwidth = 1, fill = "lightblue", color = "black", alpha = 0.7) +
  geom_density(data = df_ex4_data, aes(x = Valor), color = "darkblue", size = 1.2) +
  geom_vline(xintercept = x_bar_4, linetype = "dashed", color = "red", size = 1) +
  geom_segment(data = df_ex4_ci, aes(x = Lower, xend = Upper, y = 0, yend = 0),
               color = "darkgreen", size = 2, lineend = "round") +
  geom_point(data = df_ex4_ci, aes(x = Lower, y = 0), shape = "|", size = 8, color = "darkgreen") +
  geom_point(data = df_ex4_ci, aes(x = Upper, y = 0), shape = "|", size = 8, color = "darkgreen") +
  labs(title = "Exercício 4: Distribuição dos Dados e IC para a Média (95%)",
       x = "Valor", y = "Densidade") +
  theme_minimal() +
  annotate("text",
           x = mean(c(ci_lower_4, ci_upper_4)),
           y = -0.015,
           label = sprintf("IC (95%%): (%.2f, %.2f)", ci_lower_4, ci_upper_4),
           color = "darkgreen", size = 4) +
  coord_cartesian(clip = "off") +
  theme(plot.margin = unit(c(1, 1, 2, 1), "lines"))
```

#### **Exercício 5**

Uma marca particular de margarina diet foi analisada para determinar o nível em porcentagem de ácidos graxos insaturados. Uma amostra de seis pacotes resultou nos seguintes dados: 16,8; 17,2; 17,4; 16,9; 16,5 e 17,1. Encontre o intervalo de confiança de 99% para a amostra.

**Como a variância populacional é desconhecida e o tamanho da amostra é pequeno utilizamos a distribuição t de Student para calcular o intervalo de confiança para a média populacional.**

```r
cat("\n--- Exercício 5 ---\n")

sample_data_5 <- c(16.8, 17.2, 17.4, 16.9, 16.5, 17.1)

# Cálculo da média amostral
x_bar_5 <- mean(sample_data_5)
# Cálculo do desvio padrão amostral (ddof=1 é o padrão para sd() no R)
s_5 <- sd(sample_data_5)
# Tamanho da amostra
n_5 <- length(sample_data_5)
# Nível de confiança desejado
conf_level_5 <- 0.99
# Graus de liberdade (n-1 para t-student)
df_5 <- n_5 - 1

# Cálculo do valor crítico t para o nível de confiança
alpha_5 <- 1 - conf_level_5
t_alpha_2_5 <- qt(1 - alpha_5 / 2, df_5)

# Cálculo da margem de erro
margin_error_5 <- t_alpha_2_5 * (s_5 / sqrt(n_5))

# Cálculo dos limites do Intervalo de Confiança (IC)
ci_lower_5 <- x_bar_5 - margin_error_5
ci_upper_5 <- x_bar_5 + margin_error_5


cat(sprintf("Intervalo de Confiança (99%%): (%.3f, %.3f)\n\n", ci_lower_5, ci_upper_5))


#Gráfico
df_ex5_data <- data.frame(Valor = sample_data_5)
df_ex5_ci <- data.frame(
  Medida = "Média Gordura Insaturada",
  Estimativa = x_bar_5,
  Lower = ci_lower_5,
  Upper = ci_upper_5
)

ggplot() +
  geom_point(data = df_ex5_data, aes(x = 1, y = Valor),
             position = position_jitter(width = 0.05),
             size = 3, color = "darkred", alpha = 0.7) +
  geom_point(data = df_ex5_ci, aes(x = 1, y = Estimativa),
             size = 5, color = "blue", shape = 18) +
  geom_errorbar(data = df_ex5_ci, aes(x = 1, ymin = Lower, ymax = Upper),
                width = 0.05, color = "blue", size = 1.2) +
  geom_hline(yintercept = x_bar_5, linetype = "dashed", color = "red", size = 0.8) +
  labs(title = "Exercício 5: Dados Amostrais e IC para Média (99%)",
       y = "Porcentagem de Ácidos Graxos", x = "") +
  theme_minimal() +
  scale_x_continuous(breaks = NULL) +
  coord_cartesian(ylim = c(min(sample_data_5) - 0.5, max(sample_data_5) + 0.5)) +
  annotate("text", x = 1, y = min(sample_data_5) - 0.7,
           label = sprintf("IC (99%%): (%.2f, %.2f)", ci_lower_5, ci_upper_5),
           color = "blue", size = 4) +
  theme(plot.margin = unit(c(1, 1, 2, 1), "lines"))
```

#### **Exercício 6**

Uma amostra piloto com 12 elementos tece média de 6,7 e desvio padrão de 1,7. Qual deve ser o tamanho da amostra para que a semi amplitude do intervalo de 99,5% de confiança da média populacional não seja superior a 0,8

**Como a variância populacional é desconhecida e estamos estimando o tamanho da amostra com base em uma amostra piloto, utilizamos a distribuição t de Student**

```r
cat("--- Exercício 6 ---\n")

# Dados fornecidos
s_6 <- 1.7
E_desejado_6 <- 0.8
conf_level_6 <- 0.995
n_piloto <- 12

# Cálculo do valor crítico t
alpha_6 <- 1 - conf_level_6
t_alpha_2_6 <- qt(1 - alpha_6 / 2, df = n_piloto - 1)

# Cálculo do tamanho da amostra necessário
n_6 <- ((t_alpha_2_6 * s_6) / E_desejado_6)^2
n_6_rounded <- ceiling(n_6)  # Arredondar para o próximo inteiro

cat(sprintf("Tamanho da amostra necessário: %.0f\n\n", n_6_rounded))


#Gráfico
n_range <- seq(10, max(200, n_6_rounded + 50), by = 1)

df_plot_6 <- data.frame(N = n_range) %>%
  mutate(Semi_Amplitude_Calculada = (t_alpha_2_6 * s_6) / sqrt(N))

ggplot(df_plot_6, aes(x = N, y = Semi_Amplitude_Calculada)) +
  geom_line(color = "darkblue", size = 1) +
  geom_point(color = "darkblue", size = 2) +
  geom_hline(yintercept = E_desejado_6, linetype = "dashed", color = "red", size = 1) +
  geom_vline(xintercept = n_6_rounded, linetype = "dashed", color = "darkgreen", size = 1) +
  annotate("point", x = n_6_rounded, y = E_desejado_6, color = "purple", size = 4, shape = 18) +
  annotate("text",
           x = max(n_range) * 0.7,
           y = E_desejado_6 + 0.05,
           label = sprintf("Semi-amplitude desejada: %.1f", E_desejado_6),
           color = "red", size = 4) +
  annotate("text",
           x = n_6_rounded + 15,
           y = max(df_plot_6$Semi_Amplitude_Calculada) * 0.8,
           label = sprintf("N necessário: %.0f", n_6_rounded),
           color = "darkgreen", size = 4, hjust = 0) +
  labs(
    title = "Exercício 6: Relação entre Tamanho da Amostra (N) e Semi-Amplitude",
    subtitle = sprintf("Para σ estimado = %.1f e Nível de Confiança = %.1f%%", s_6, conf_level_6 * 100),
    x = "Tamanho da Amostra (N)",
    y = "Semi-Amplitude do Intervalo de Confiança"
  ) +
  theme_minimal() +
  coord_cartesian(clip = "off") +
  theme(plot.margin = unit(c(1, 1, 1, 1), "lines"))
```

#### **Exercício 7**

Em uma amostra aleatória de 85 mancais de eixos de manivelas de motores de automóveis, 10 têm um acabamento superficial mais rugoso do que as especificações permitidas. Calcule um intervalo de confiança para o 95% da proporção

**Como estamos estimando uma proporção populacional com base em uma amostra aleatória e o tamanho amostral é suficientemente grande, utilizamos a distribuição normal padrão (z)**

```r
cat("--- Exercício 7 ---\n")
x_7 <- 10
n_7 <- 85
p_hat_7 <- x_7 / n_7
conf_level_7 <- 0.95
alpha_7 <- 1 - conf_level_7
z_alpha_2_7 <- qnorm(1 - alpha_7 / 2)
margin_error_7 <- z_alpha_2_7 * sqrt((p_hat_7 * (1 - p_hat_7)) / n_7)
ci_lower_7 <- p_hat_7 - margin_error_7
ci_upper_7 <- p_hat_7 + margin_error_7

cat(sprintf("Intervalo de Confiança (95%%): (%.3f, %.3f)\n\n", ci_lower_7, ci_upper_7))

# Gráfico
df_ex7 <- data.frame(
  Categoria = "Mancais Rugosos",
  Proporcao = p_hat_7,
  Lower = ci_lower_7,
  Upper = ci_upper_7
)

ggplot(df_ex7, aes(x = Categoria, y = Proporcao)) +
  geom_bar(stat = "identity", fill = "salmon", width = 0.4) +
  geom_errorbar(aes(ymin = Lower, ymax = Upper), width = 0.1, color = "darkred", size = 1.2) +
  geom_point(size = 4, color = "darkred") +
  labs(title = "Exercício 7: IC para Proporção de Mancais Rugosos (95%)",
       y = "Proporção", x = "") +
  theme_minimal() +
  scale_y_continuous(labels = scales::percent_format(accuracy = 1)) +
  coord_cartesian(ylim = c(0, max(ci_upper_7) * 1.2)) # Ajusta o zoom do eixo Y
```

#### **Exercício 8**

De 1.000 casos selecionados de aleatoriamente de câncer de pulmão, 823 resultaram em morte. Construa um intervalo de confiança de 95% para a taxa de morte de câncer de pulmão.

**Como estamos estimando uma proporção populacional com base em uma amostra aleatória e de grande tamanho, utilizamos a distribuição normal padrão (z) para calcular o intervalo de confiança**

```r
cat("--- Exercício 8 ---\n")

# Dados do exercício
x_8 <- 823
n_8 <- 1000
conf_level_8 <- 0.95

# Cálculo manual do intervalo de confiança
p_hat_8 <- x_8 / n_8
z_alpha_2_8 <- qnorm(1 - (1 - conf_level_8) / 2)
erro_8 <- z_alpha_2_8 * sqrt(p_hat_8 * (1 - p_hat_8) / n_8)
ci_manual <- c(p_hat_8 - erro_8, p_hat_8 + erro_8)

cat(sprintf("IC (95%%, manual): (%.3f, %.3f)\n", ci_manual[1], ci_manual[2]))

# IC com prop.test (sem correção de continuidade)
ic_prop <- prop.test(x_8, n_8, conf.level = conf_level_8, correct = FALSE)$conf.int
cat(sprintf("IC (95%%, prop.test): (%.3f, %.3f)\n\n", ic_prop[1], ic_prop[2]))

# Dataframe para o gráfico
df_ex8_plot <- data.frame(
  Medida = "Taxa de Morte (Câncer de Pulmão)",
  Proporcao = p_hat_8,
  Lower = ci_manual[1],
  Upper = ci_manual[2]
)

# Gráfico
ggplot(df_ex8_plot, aes(x = Medida, y = Proporcao)) +
  geom_col(fill = "lightcoral", width = 0.4) +
  geom_errorbar(aes(ymin = Lower, ymax = Upper), width = 0.1, color = "darkred", size = 1.2) +
  geom_point(size = 4, color = "darkred", shape = 18) +
  geom_text(aes(label = sprintf("%.1f%%", Proporcao * 100)), vjust = -0.8, color = "darkred", size = 4) +
  geom_text(aes(y = min(0.01, Lower * 0.9),
                label = sprintf("IC: %.1f%% - %.1f%%", Lower * 100, Upper * 100)),
            vjust = 2, color = "darkred", size = 3.5) +
  labs(
    title = "IC para Taxa de Morte por Câncer de Pulmão (95%)",
    x = NULL, y = "Proporção"
  ) +
  scale_y_continuous(
    labels = scales::percent_format(accuracy = 0.1),
    limits = c(0, df_ex8_plot$Upper * 1.1)
  ) +
  coord_cartesian(clip = "off") +
  theme_minimal() +
  theme(plot.margin = unit(c(1, 1, 2, 1), "lines"))
```

Pelos resultados dos intervalos de confiança, vemos que, em termos práticos, os dois métodos de cálculo são praticamente equivalentes. A diferença entre eles é mínima e não altera a interpretação dos dados

#### **Exercício 9**

Sabe-se que a proporção de animais contaminados com uma determinada doença não é superior a 10%. Qual deve o tamanho da amostra para que a semi amplitude do intervalo com 92% de confiança para a fração populacional não seja superior a 2%?

**Como a proporção populacional é estimada em até 10% e desejamos um intervalo com 92% de confiança e erro máximo de 2%, usamos a distribuição normal padrão (z) para calcular o tamanho mínimo da amostra**

```r
cat("--- Exercício 9 ---\n")

# Parâmetros do exercício
E_9 <- 0.02
conf_level_9 <- 0.92
z_alpha_2_9 <- qnorm(1 - (1 - conf_level_9) / 2)

# Estimativa otimista (p = 0.10)
p_optimistic_9 <- 0.10
n_optimistic_9 <- ceiling((z_alpha_2_9 / E_9)^2 * p_optimistic_9 * (1 - p_optimistic_9))

# Estimativa conservadora (p = 0.50)
p_conservative_9 <- 0.5
n_conservative_9 <- ceiling((z_alpha_2_9 / E_9)^2 * p_conservative_9 * (1 - p_conservative_9))

cat("\n--- Tamanhos de Amostra Necessárop ---\n")
cat(sprintf("Estimativa Otimista (p = %.2f): %d\n", p_optimistic_9, n_optimistic_9))
cat(sprintf("Estimativa Conservadora (p = %.2f): %d\n\n", p_conservative_9, n_conservative_9))


#Gráfico

# Gera sequência de proporções e calcula tamanho da amostra para cada valor
df_plot_9 <- data.frame(p = seq(0.01, 0.99, 0.01)) %>%
  mutate(N_necessario = ceiling((z_alpha_2_9 / E_9)^2 * p * (1 - p)))

# Pontos de destaque
df_destaques <- data.frame(
  p = c(p_optimistic_9, p_conservative_9),
  N = c(n_optimistic_9, n_conservative_9),
  Tipo = c("Otimista (p = 10%)", "Conservador (p = 50%)")
)

ggplot(df_plot_9, aes(x = p, y = N_necessario)) +
  geom_line(color = "darkblue", size = 1) +
  geom_point(color = "darkblue", size = 2) +
  geom_point(data = df_destaques, aes(x = p, y = N, color = Tipo), size = 6, shape = 18) +
  geom_text(data = df_destaques, aes(x = p, y = N, label = sprintf("N = %d", N), color = Tipo),
            vjust = -1.5, size = 4) +
  geom_hline(yintercept = n_optimistic_9, linetype = "dotted", color = "darkgreen", size = 0.8) +
  geom_hline(yintercept = n_conservative_9, linetype = "dotted", color = "darkred", size = 0.8) +
  labs(
    title = "Exercício 9: Tamanho da Amostra em Função da Proporção (p)",
    subtitle = sprintf("Para margem de erro de %.0f%% e %.1f%% de confiança", E_9 * 100, conf_level_9 * 100),
    x = "Proporção Estimada (p)",
    y = "Tamanho da Amostra Necessário (N)"
  ) +
  scale_x_continuous(labels = scales::percent_format(accuracy = 1)) +
  scale_y_continuous(labels = scales::comma) +
  scale_color_manual(values = c(
    "Conservador (p = 50%)" = "darkred",
    "Otimista (p = 10%)" = "darkgreen"
  )) +
  theme_minimal() +
  theme(
    legend.position = "bottom",
    plot.margin = unit(c(1, 1, 1, 1), "lines")
  )
```
